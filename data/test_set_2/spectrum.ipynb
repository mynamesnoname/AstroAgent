```python
import os 
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
```

```python
fits_path = 'all_spectra_per_wave.fits'
with fits.open(fits_path) as hdul:
    meta = hdul['METADATA'].data
    spec = hdul['SPECTRA'].data
    print(f'How many pieces of spectrum: {len(spec)}')
```

```python
print("spec的列:", spec.dtype.names)
```

```python
plt.figure(figsize = (12, 4))
i = 0 # That's 001.png
plt.plot(spec['WAVE_B'][i], spec['FLUX_B'][i], color='k', lw=0.5, alpha=0.8)
plt.plot(spec['WAVE_R'][i], spec['FLUX_R'][i], color='b', lw=0.5, alpha=0.8)
plt.plot(spec['WAVE_Z'][i], spec['FLUX_Z'][i], color='r', lw=0.5, alpha=0.8)
plt.xlabel('Wavelength [Angstrom]')
plt.ylabel('Flux [1e-17 erg/s/cm2/Angstrom]')
```

```python
plt.figure(figsize = (12, 4))
i = 0 # That's 001.png
sigma = 4 # Gaussian filter width
plt.plot(spec['WAVE_B'][i], gaussian_filter1d(spec['FLUX_B'][i], sigma), color='k', lw=0.5, alpha=0.8)
plt.plot(spec['WAVE_R'][i], gaussian_filter1d(spec['FLUX_R'][i], sigma), color='b', lw=0.5, alpha=0.8)
plt.plot(spec['WAVE_Z'][i], gaussian_filter1d(spec['FLUX_Z'][i], sigma), color='r', lw=0.5, alpha=0.8)
plt.xlabel('Wavelength [Angstrom]')
plt.ylabel('Flux [1e-17 erg/s/cm2/Angstrom]')
```
